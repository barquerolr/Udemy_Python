```python
def user_choice():
    #TWO VARIABLES
    
    #INITIAL VARIABLES
    
    choice = 'WRONG'
    acceptable_values = range(0, 10+1)
    within_range = False
    while choice.isdigit() == False or within_range == False:
        #DIGIT CHECK
        choice = input("Plase enter a number between (0 and 10): ")
        print("Invalid Entry. Not a number")
        
        if choice.isdigit():
            
            #Range Check
            if int(choice) in acceptable_values:
                within_range = True
                
            else:
                within_range = False
                print("Out of range of 0 and 10")
                
    return int(choice)
```
